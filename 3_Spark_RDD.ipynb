```python
from pyspark import SparkConf, SparkContext
```

```python
if __name__ == "__main__":
  conf = SparkConf().setAppName("Tushar").setMaster("local[3]")
  sc = SparkContext(conf = conf)
```

```python
data = sc.textFile("retaildir/retailstore_large.csv")
```

```python
data.getNumPartitions()
```

```python
data.count()
```

# Show data

```python
data.top(5)
```

```python
data.take(3)
```

# Flat map

```python
result = data.flatMap(lambda x:x.split(","))
result.take(20)
```

```python
result.coalesce(2).saveAsTextFile("result.csv")
```

```python
result.repartition(10).saveAsTextFile("newresult.csv")
```
